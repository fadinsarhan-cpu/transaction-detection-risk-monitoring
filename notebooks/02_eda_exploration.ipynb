# 02_EDA: Exploratory Data Analysis

This notebook explores the cleaned transaction dataset to understand class imbalance, amount distribution, transaction characteristics, and key drivers of suspicious activity.


## Setup and Load Cleaned Data

Import the cleaned dataset for exploratory data analysis.

```

from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styles
sns.set(style='whitegrid')

DATA_PROCESSED_DIR = Path('data/processed')
df = pd.read_csv(DATA_PROCESSED_DIR / 'cleaned_transactions.csv')
print('Data loaded with shape', df.shape)
```

## Class Imbalance Analysis

Examine the distribution of the target variable `Is_laundering` to understand class imbalance.

```

# Plot class distribution
if 'is_laundering' in df.columns:
    fig, ax = plt.subplots(figsize=(6,4))
    counts = df['is_laundering'].value_counts()
    counts.plot(kind='bar', ax=ax)
    ax.set_title('Class Distribution of Suspicious Transactions')
    ax.set_xlabel('Is_laundering')
    ax.set_ylabel('Count')
    plt.show()
    # Print imbalance ratio
    ratio = counts.min() / counts.max()
    print('Imbalance ratio (minority/majority):', ratio)
```

## Transaction Amount Distribution

Visualize the distribution of transaction amounts to identify outliers and general patterns.

```

if 'amount' in df.columns:
    fig, ax = plt.subplots(figsize=(6,4))
    sns.histplot(df['amount'], bins=50, kde=True, ax=ax)
    ax.set_title('Transaction Amount Distribution')
    ax.set_xlabel('Amount')
    plt.show()
    # Boxplot by class
    if 'is_laundering' in df.columns:
        plt.figure(figsize=(6,4))
        sns.boxplot(data=df, x='is_laundering', y='amount')
        plt.title('Amount Distribution by Class')
        plt.show()
```

## Suspicious Transactions by Payment Type

Identify the distribution of suspicious transactions across different payment types.

```

if {'payment_type','is_laundering'}.issubset(df.columns):
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x='payment_type', hue='is_laundering')
    plt.xticks(rotation=45)
    plt.title('Suspicious Transactions by Payment Type')
    plt.xlabel('Payment Type')
    plt.ylabel('Count')
    plt.show()
```

## Suspicious Transactions by Currency

Explore how suspicious activity varies across different currencies.

```

if {'currency','is_laundering'}.issubset(df.columns):
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x='currency', hue='is_laundering')
    plt.xticks(rotation=45)
    plt.title('Suspicious Transactions by Currency')
    plt.xlabel('Currency')
    plt.ylabel('Count')
    plt.show()
```

## Temporal Patterns of Suspicious Transactions

If a datetime column exists, analyze the frequency of suspicious transactions over time.

```

if {'datetime','is_laundering'}.issubset(df.columns):
    df['date'] = pd.to_datetime(df['datetime']).dt.date
    suspicious_by_date = df[df['is_laundering']==1].groupby('date').size()
    plt.figure(figsize=(10,4))
    suspicious_by_date.plot()
    plt.title('Suspicious Transactions Over Time')
    plt.xlabel('Date')
    plt.ylabel('Count of Suspicious Transactions')
    plt.show()
```

## Top Accounts by Suspicious Activity

Find sender or receiver accounts involved in the highest number of suspicious transactions.

```

if {'sender_account','is_laundering'}.issubset(df.columns):
    top_senders = df[df['is_laundering']==1].groupby('sender_account').size().sort_values(ascending=False).head(10)
    print('Top sender accounts with suspicious activity:')
    display(top_senders)
if {'receiver_account','is_laundering'}.issubset(df.columns):
    top_receivers = df[df['is_laundering']==1].groupby('receiver_account').size().sort_values(ascending=False).head(10)
    print('Top receiver accounts with suspicious activity:')
    display(top_receivers)
```
